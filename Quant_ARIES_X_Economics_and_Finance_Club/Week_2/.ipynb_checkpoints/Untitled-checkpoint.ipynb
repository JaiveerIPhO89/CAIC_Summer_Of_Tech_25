## **WEEK 2 CHALLENGE :**  *Stock Market Prediction*
* In this task, we will analyse, train on and predict the stock market exchange values using three predictive models :
    1. *Linear Regression*
    2. *ARIMA*
    3. *Random Forest*
* The accuracy of our models will be judged on the **Mean Square Error (MSE)** and **Direction Prediction Accuracy** over the entire test dataset.
* The following quantities will be additionally calculated, tabulated and compared :
| Quantity | Definition |
| -------- | ---------- |
| $7$-day Moving Average | The average closing price over the span of the last week |
| $30$-day Moving Average | The average closing price over the span of the last month. |
| Relative Strength Index | The fraction $ 100 * \dfrac{Average Loss}{Average Loss + Average Gain} $ over the span of last 14 days. |

```python
%matplotlib inline

import pandas as pd
import numpy as np

from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from sklearn.ensemble import RandomForestRegressor

import os
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns

sns.set(style="darkgrid", context="talk", palette="Set2")
```

### **STEP 1 :** *Preparing Dataset*
* Firstly, we will load the stocks exchange data of `Amazon` and assign them to a *dataframe with the datetime index*.
* Next, we will re-index the dataframe to include all the dates within the timeframe of the stocks data, and use forward fill to compute any missing entries.
* Then we will compute and add the *$7$-day Moving Average*, *$30$-day Moving Average*, and *Relative Strength Index* for every entry.
* We will also add columns for the *last $5$ closing prices*, which will be useful for *Linear Regression*.
* Lastly, we will split our dataset, the first $80%$ for *model training* and the rest $20%$ for model testing.

```python
def load_stock_data(stock_dir, ticker):
    file_path = os.path.join(stock_dir, f'{ticker}.txt')
    df = pd.read_csv(file_path)
    df['Date'] = pd.to_datetime(df['Date'])
    df = df.sort_values('Date')
    df.set_index('Date', inplace=True)
    return df

stock_dir = 'data'
ticker = 'amzn.us'

df_raw = load_stock_data (stock_dir, ticker)
```

```python
df_raw.head()
```

```python
df_raw.tail()
```

```python
def find_all_dates(df):
    start_date = df.index.min()
    end_date = df.index.max()
    all_dates = pd.date_range(start=start_date, end=end_date, freq='B')

    return all_dates

all_dates = find_all_dates(df_raw)
```

```python
def find_missing_dates(df, all_dates):
    missing_dates = pd.Series(all_dates.difference(df.index))

    if not missing_dates.empty:
        print("The missing dates are :")
        print(missing_dates)
```

```python
find_missing_dates(df_raw, all_dates)
```

```python
def pad_stocks_data(df, all_dates):
    df = df.reindex(all_dates)
    df.index.name = 'Date'
    df = df.ffill()
    return df

df_clean = pad_stocks_data(df_raw, all_dates)
```

```python
df_clean.head()
```

```python
df_clean.tail()
```

```python
def add_features(df_clean):
    df = df_clean.copy()

    df['Daily Return'] = df['Close'].pct_change(fill_method=None)
    df['MA7'] = df['Close'].rolling(window=7).mean()
    df['MA30'] = df['Close'].rolling(window=30).mean()
    df['RV30'] = df['Daily Return'].transform(lambda x: x.rolling(window=30).std())
    
    # RSI (14-day)
    delta = df['Close'].diff()
    gain = np.where(delta > 0, delta, 0)
    loss = np.where(delta < 0, -delta, 0)
    roll_up = pd.Series(gain, index=df.index).rolling(14).mean()
    roll_down = pd.Series(loss, index=df.index).rolling(14).mean()
    rs = roll_up / roll_down
    df['RSI'] = 100 - (100 / (1 + rs))

    # Lag features (for Linear Regression)
    for lag in range(1, 6):
        df[f'lag_{lag}'] = df['Close'].shift(lag)
    
    return df

df_final = add_features(df_clean)
```

```python
df_final.head()
```

```python
df_final.tail()
```

```python
def split_train_test(df, split_ratio=0.8):
    n = len(df)
    split_point = int(n * split_ratio)
    
    df_train = df.iloc[:split_point].copy()
    df_test = df.iloc[split_point:].copy()
    
    return df_train, df_test

df_train, df_test = split_train_test(df_final)
```

### **STEP 2 :** *Linear Regression*
* Firstly, we will prepare the **X-variable** (*[lag_1, lag_2, lag_3, lag_4, lag_5]*) and the **Y-variable** (*close*) for both the *training and testing datasets*.
* Next, we will open the *Linear Regression Model*, *train it* on the training dataset, and gather it's *predictions* on the testing dataset.
* Then, we will evaluate the *Mean Square Error* and the *Direction Prediction Accuracy* to judge the performance of our model.
* Lastly, we will plot the *Actual & Predicted Prices*, *Residual Errors* and *Direction Predictions* vs *Datetime* to visualize the performance.

```python
def prepare_lr_data(df):
    df = df.copy()
    
    feature_cols = [f'lag_{i}' for i in range(1, 6)]     # Features: lag_1 to lag_5
    
    df = df.dropna(subset=feature_cols + ['Close'])      # Drop rows with NaNs (e.g. first few rows due to lag)
    
    X = df[feature_cols].values
    y = df['Close'].values
    
    return X, y

X_train, y_train = prepare_lr_data(df_train)
X_test, y_test = prepare_lr_data(df_test)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

y_pred = lr_model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

actual_direction = (y_test[1:] > y_test[:-1])
predicted_direction = (y_pred[1:] > y_test[:-1])
direction_accuracy = (actual_direction == predicted_direction).mean()

print(f'Linear Regression MSE: {mse:.4f}')
print(f'Linear Regression RMSE: {rmse:.4f}')
print(f"Direction Prediction Accuracy: {direction_accuracy:.2%}")


```

```python
plt.figure(figsize=(14, 6))
plt.plot(df_test.index, y_test, label='Actual Price', color='black')
plt.plot(df_test.index, y_pred, label='Predicted Price', color='green', linestyle='--')
plt.title('Linear Regression: Actual vs Predicted Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
residuals = y_test - y_pred

plt.figure(figsize=(14, 4))
plt.plot(df_test.index, residuals, color='red')
plt.axhline(0, linestyle='--', color='black')
plt.title("Residuals Over Time")
plt.xlabel("Date")
plt.ylabel("Error (Actual - Predicted)")
plt.show()
```

```python
# Calculate direction correctness
actual_direction = np.sign(np.diff(y_test))
predicted_direction = np.sign(np.diff(y_pred))
correct = actual_direction == predicted_direction

# Plot
plt.figure(figsize=(14, 4))
plt.scatter(df_test.index[1:], correct, c=correct, cmap='bwr', edgecolor='k', label='Direction Accuracy')
plt.title("Direction Prediction Accuracy (Scatter)")
plt.xlabel("Date")
plt.ylabel("Correct (1) / Incorrect (0)")
plt.yticks([0, 1])
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.show()
```

### **RESULTS**
* The Regression Model is *brilliantly* able to predict the stock prices at any date, with a *RMSE loss* of $\sim 10.25\$$.
* However, the *Direction Prediction* is as good as *flipping a coin*, as the accuracy turns out to be $\sim 52.95\%$.

### **STEP 3 :** *ARIMA*
* It turns out that *ARIMA modelling* is not a good method for this stock dataset.
* The analysis of *Autocorrelation* and *Partial Autocorrelation* shows us that there is virtually no standout value of $p$ and $q$ suitable for the model.
* Even the modelling with $ p, q, d = 2, 1, 2$ shows us that the model isn't able to pick up on the stock price movement during the testing.

```python
def adf_test(series):
    result = adfuller(series)
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    if result[1] <= 0.05:
        print(" Series is stationary")
    else:
        print(" Series is NOT stationary")
```

```python
adf_test(df_train['Close'])
```

```python
train_diff = df_train['Close'].diff().dropna()

adf_test(train_diff)
```

```python
plt.figure(figsize=(12, 4))
plot_acf(train_diff, lags=40)
plt.show()

plt.figure(figsize=(12, 4))
plot_pacf(train_diff, lags=40)
plt.show()
```

```python
# 1. Extract training and test series
train_series = df_train['Close']
test_series = df_test['Close']
```

```python
# 2. Fit ARIMA model (basic parameters)
model = ARIMA(train_series, order=(2, 1, 2))
model_fit = model.fit()
```

```python
# 3. Forecast
forecast = model_fit.forecast(steps=len(test_series))
forecast.index = test_series.index
```

```python
mse = mean_squared_error(test_series, forecast)
rmse = np.sqrt(mse)

print(f'ARIMA MSE: {mse:.4f}')
print(f'ARIMA RMSE: {rmse:.4f}')
```

```python
# 5. Plot
plt.figure(figsize=(14, 6))
plt.plot(train_series.index, train_series, label='Train')
plt.plot(test_series.index, test_series, label='Actual Test')
plt.plot(forecast.index, forecast, label='Forecast', linestyle='--')
plt.title('ARIMA Forecast vs Actual')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```
